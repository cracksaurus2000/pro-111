```python
# Mostrar imagen
from matplotlib import pyplot
from matplotlib.image import imread

training_damaged_image = "/content/PROC111-V1-plantilla-proyecto/train/damage/image (1).jpeg"

# Cargar los pixeles de la imagen
image = imread(training_damaged_image)

pyplot.title("Da√±ado: Imagen 1")

# Graficar los datos de los pixeles sin procesar
pyplot.imshow(image)

# Mostrar la figura
pyplot.show()
```

```python
# Procesando rotaci√≥n de imagen (la actividad del alumno comienza aqu√≠)
import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Rotaci√≥n aleatoria usando Image Data Generator 
training_data_generator = ImageDataGenerator(rotation_range=90, fill_mode='nearest')

# Directorio de las im√°genes
training_image_directory = "/content/PROC111-V1-plantilla-proyecto/train"

# Generando archivos de imagen rotados aleatoriamente
training_augmented_images = training_data_generator.flow_from_directory(training_image_directory,target_size=(180,180))
```

```python
# Visualizar las im√°genes preprocesadas
from matplotlib import pyplot
from matplotlib.image import imread

for i in range(4):

	# Definir una subfigura
	pyplot.subplot(2, 2, i+1)
 
	# Generar un lote de im√°genes
	batch = training_augmented_images.next()
 
	# Convertir a n√∫meros enteros sin signo para la visualizaci√≥n
	image = batch[0][0].astype('uint8')
 
	# Graficar datos de pixeles sin procesar
	pyplot.imshow(image)

# Graficar la figura principal (con 4 subfiguras)
pyplot.show()
```

```python
# Zoom e inverci√≥n horizontal

import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Generaci√≥n artificial de datos aleatoria (Reescalar, Redimensionar, Desplazar, Rotar) usando ImageDataGenerator 
training_data_generator = ImageDataGenerator(zoom_range=0.3, 
                                             horizontal_flip=True, 
                                             fill_mode='nearest')

# Directorio de im√°genes
training_image_directory = "/content/PROC111-V1-plantilla-proyecto/train"

# Archivos de imagen generados artificialmente
training_augmented_images = training_data_generator.flow_from_directory(
                                                       training_image_directory,
                                                       target_size=(180,180))
```

```python
# Visualizar imagenes preprocesadas
from matplotlib import pyplot
from matplotlib.image import imread

for i in range(4):

	# Definir una subfigura
	pyplot.subplot(2, 2, i+1)
 
	# Generar un lote de im√°genes
	batch = training_augmented_images.next()
 
	# Convertir a n√∫meros enteros sin signo para visualizaci√≥n
	image = batch[0][0].astype('uint8')
 
	# Graficar datos de pixeles sin procesar
	pyplot.imshow(image)

# Graficar la figura principal (con 4 subfiguras)
pyplot.show()
```

```python
# Ageneraci√≥n artificial de datos aleatoria (Reescalar, Rotar, Invertir, Zoom, Desplazar) usando ImageDataGenerator 
training_data_generator = ImageDataGenerator(
    rescale = 1.0/255,
    rotation_range=40,
    width_shift_range=0.3,
    height_shift_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest')
```
